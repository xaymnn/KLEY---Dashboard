```python
import csv
import random
from datetime import datetime, timedelta

# Liste des résidences KLEY
residences = [
    "Angers",
    "Brest",
    "Caen",
    "Cergy",
    "Champs",
    "Marseille",
    "Montpellier",
    "Paris",
    "Reims",
    "Rennes",
    "Rouen",
    "Saclay",
    "Toulouse"
]

# Liste des types d'écoles
types_ecoles = [
    "Université",
    "École de Commerce",
    "École Spécialisée",
    "IUT",
    "BTS"
]

# Liste des types de logements

types_logements = [
    "KOCOON",
    "KOCOON L",
    "KOCOON XL",
    "KABANE",
    "KASA"
]

# Générer une date aléatoire dans une période donnée
def random_date(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return start_date + timedelta(days=random_days)

# Générer le jeu de données
data = []
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 12, 31)

for residence in residences:
    for _ in range(random.randint(50, 125)):
        ecole = random.choice(types_ecoles)
        logement = random.choice(types_logements)
        age = random.randint(18, 25)
        arrivee_date = random_date(start_date, end_date)
        revenu_mensuel = random.randint(1000, 2000)
        niveau_satisfaction = random.choices(range(11), weights=[0, 1, 2, 2, 3, 3, 3, 3, 4, 3, 2])[0]

        data.append([
            residence,
            ecole,
            logement,
            age,
            arrivee_date.strftime("%Y-%m-%d"),
            revenu_mensuel,
            niveau_satisfaction
        ])

# Afficher le jeu de données généré
for entry in data:
    print(entry)

# Définir le nom du fichier CSV de sortie
filename = "donnees_residences.csv"

# Écrire les données dans le fichier CSV en spécifiant l'encodage
with open(filename, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    # Écrire l'en-tête du fichier CSV
    writer.writerow(["Résidence", "École", "Type de logement", "Âge", "Date d'arrivée", "Revenu mensuel", "Niveau de satisfaction"])
    # Écrire les lignes de données
    writer.writerows(data)

print("Le fichier CSV a été généré avec succès.")
```
